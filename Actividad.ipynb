### Librerias necesarias:

```python
import json
import requests
import pandas as pd
```

### API de  lugares

```python
url:str = r'https://fakerapi.it/api/v1/places?_quantity=100'
```

```python
response:object = requests.get(url)
```

```python
pd.DataFrame(json.loads(response.content.decode()).get('data'))
```

### API de Imagenes

```python
url:str = r'https://fakerapi.it/api/v1/images?_width=380?_quantity=100'
```

```python
response:object = requests.get(url)
```

```python
pd.DataFrame(json.loads(response.content.decode()).get('data'))
```

### API de Users

```python
url:str = r'https://fakerapi.it/api/v1/users?_quantity=400'
```

```python
response:object = requests.get(url)
```

```python
pd.DataFrame(json.loads(response.content.decode()).get('data'))
```
